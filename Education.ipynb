## Data Visualization

To indicate the scatter plot of poopulation who are go study in high school along with the number of population who complete highschool and the average year to complete

```python
import seaborn as sns
plt.subplots(figsize=(12, 8))
sns.set(font_scale=1.2)
ax = sns.scatterplot(x="CompleteHigh", y="AvgYearHigh", hue='HighSchool', data=df)
```

Using heatmap to generate the correlation of two attributes. The red level show the positive correlation, while the green level shows the unrelative relation between two attributes. Following this map, it indicates that people who attends in secondary school and high school has trend to graduate from each education level, while student who attends primary school has the  less trend to go up in each education level and complete them.

```python
import seaborn as sns
import numpy as np

fi, axis = plt.subplots(figsize=(10, 8))
corr = df.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(200,10,as_cmap=True), square=True, ax=axis)
```

To show the distribution for population in each year.

```python
plt.hist(df.population, bins='auto') 
plt.title("The histogram showing the distribution of population")
plt.xlabel('population')
plt.ylabel('frequencies')
plt.show()
```

```python
sns.distplot(df.population)
plt.title("The histogram showing the distribution of population")
```

To diaplay the distribution of the population who no attend to school in each year comparing with the population who attend to school for each year. Following this two graph, the population of no-school people is fluctutation, while the population for going school people has a trend to in crease in each year.

```python
fi= plt.figure(figsize=(14,5))

axis=fi.add_subplot(121)
sns.distplot(df[(df.NoSchool==1)]['year'],color='b',ax=axis)
axis.set_title('The distribution of the population for the no-school people')

axis=fi.add_subplot(122)
sns.distplot(df[(df.NoSchool ==0)]['year'],color='c',ax=axis)
axis.set_title('The distribution of the population for people who go to school')
```

To display the outlier of complete high school for each country in region group based on region code and in each year. Following the box, there are more outlier in increasing year for each region code. 

```python
sns.catplot(x='year', y='region_code', data=df[(df.CompleteHigh ==0.0)] , orient='h', kind='box' , palette = 'inferno', height=10, aspect=2)
plt.title("The box plot of complete high school for each year and each region based on region code")
```
